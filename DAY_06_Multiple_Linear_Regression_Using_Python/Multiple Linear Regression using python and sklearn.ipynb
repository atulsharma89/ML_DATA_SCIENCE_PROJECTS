```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
dataset = pd.read_csv('50_Startups.csv')

dataset.head()
```

```python
X=dataset.iloc[:,:-1]

X.head()
```

```python
y=dataset.iloc[:,4]

y.head()
```

```python
states=pd.get_dummies(X['State'],drop_first=True)

states
```

```python
X=X.drop('State',axis=1)
```

```python
X=pd.concat([X,states], axis = 1)

X.head()
```

```python
#splitting dataset into trainning and testing

from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)

X_train.head()
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train,y_train)

```

```python
#prediction on y_test

y_pred=regressor.predict(X_test)

y_pred
```

```python
#r-squate = 1 - (SSres/SSmean)

#SSres = (1/n)(summation (y-y_pred)^2)

#SSmean = (1/n)(summation (y-y_mean)^2)

#3ssmean> ssres  is our model is good

# 1 - SMALL NUMBER IS alomost equal to 0.9

# more nearer to 1 explains that our model is good
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test,y_pred)
```
